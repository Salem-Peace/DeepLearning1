```python
import os
import random
import shutil

# Define the path to the original image folders and the destination folder for reduced data
original_data_dir = '/Users/temitopeagbabiaka/Downloads/archive (3)/train/'
destination_dir = '/Users/temitopeagbabiaka/Downloads/reductional/'

# Determine the number of images to keep (approximately 50% reduction)
reduction_factor = 0.5

# Iterate over each folder containing images
for category in os.listdir(original_data_dir):
    category_dir = os.path.join(original_data_dir, category)
    if os.path.isdir(category_dir):
        # Get the list of image filenames in the folder
        image_filenames = os.listdir(category_dir)
        # Shuffle the list of image filenames
        random.shuffle(image_filenames)
        # Calculate the number of images to keep
        num_images_to_keep = int(len(image_filenames) * reduction_factor)
        # Select the first num_images_to_keep filenames from the shuffled list
        selected_filenames = image_filenames[:num_images_to_keep]
        # Create the destination folder if it doesn't exist
        destination_category_dir = os.path.join(destination_dir, category)
        os.makedirs(destination_category_dir, exist_ok=True)
        # Copy or move the selected images to the destination folder
        for filename in selected_filenames:
            src = os.path.join(category_dir, filename)
            dst = os.path.join(destination_category_dir, filename)
            # Copy or move the image file to the destination folder
            shutil.copy(src, dst)  # Use shutil.move if you want to move the files instead of copying
```

### read all images and convert them to array

```python
training_Dt = []

def create_training_Dt():
    for category in Classes:
        path = os.path.join(Datadirectory, category)
        class_num = Classes.index(category) ## 0 1, label
        for img in os.listdir(path):
            try:
                img_array= cv2.imread(os.path.join(path, img))
                new_array= cv2.resize(img_array, (img_size, img_size))
                training_Dt.append([new_array, class_num])
            except Exception as e:
                pass
```

```python
create_training_Dt()
```

```python
print(len(training_Dt))
```

```python
import random

random.shuffle(training_Dt) # shuffling so tht my deep learning doesnt learn the sequence rather show the dynamic and robust...
```

```python
# create data and labels

X= []
Y = []

for features, label in training_Dt:
    X.append(features)
    Y.append(label)
    
X = np.array(X).reshape(-1, img_size, img_size, 3)
Y = np.array(Y)
```

```python
X.shape
```

```python
X = X/255.0; #normalizing it, its important to do this before training a data(225 is dark as 0 is white)
```

### Deep Learning Model Training - Transfer Learning

```python
from tensorflow import keras
from tensorflow.keras import layers
```

```python
model = tf.keras.applications.MobileNetV2() # pre-trained model
```

```python
model.summary()
```

## Transfer Learning - Tuning, weight will start from the check point

```python
base_input = model.layers[0].input
base_output = model.layers[-2].output
```

```python
base_output
```

```python
final_output = layers.Dense(128)(base_output)
final_output = layers.Activation('relu')(final_output)
final_output = layers.Dense(64)(final_output)
final_output = layers.Activation('relu')(final_output)
final_output = layers.Dense(7, activation= 'softmax')(final_output)
```

```python
final_output
```

```python
new_model = keras.Model(inputs= base_input, outputs= final_output)
```

```python
new_model.summary()
```

```python
new_model.compile(loss= 'sparse_categorical_crossentropy', optimizer= 'adam', metrics= ['accuracy'])
```

```python
history = new_model.fit(X,Y, epochs=25) # training
```

```python
from tensorflow.keras.models import load_model
from sklearn.metrics import confusion_matrix, classification_report
import seaborn as sb
from sklearn.model_selection import train_test_split

# Load your pre-trained model
model_path = 'Final_model1.h5'
new_model = load_model(model_path)

# Split the data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42)

# Ensure the model is trained
history = new_model.fit(X_train, y_train, epochs=25, validation_split=0.2)

# Make predictions on the test set
y_pred_probs = new_model.predict(X_test)
y_pred = np.argmax(y_pred_probs, axis=1)

# Define class labels
class_labels = ["Angry", "Disgust", "Fear", "Happy", "Neutral", "Sad", "Surprise"]

# Generate the confusion matrix
conf_matrix = confusion_matrix(y_test, y_pred)

# Plot the confusion matrix
plt.figure(figsize=(10, 8))
sb.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', xticklabels=class_labels, yticklabels=class_labels)
plt.xlabel('Predicted')
plt.ylabel('True')
plt.title('Confusion Matrix')
plt.show()

# Print classification report
print(classification_report(y_test, y_pred, target_names=class_labels))
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.models import load_model

# Load the pre-trained Haar Cascade classifier for face detection
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')

# Load your pre-trained model
model_path = 'Final_model1.h5'
new_model = load_model(model_path)

# Define class labels
class_labels = ["Angry", "Disgust", "Fear", "Happy", "Neutral", "Sad", "Surprise"]

def process_faces(image):
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    faces = face_cascade.detectMultiScale(gray, 1.1, 4)
    
    face_rois = []
    coords = []
    
    for x, y, w, h in faces:
        roi_color = image[y:y+h, x:x+w]
        face_rois.append(roi_color)
        coords.append((x, y, w, h))

    return face_rois, coords

def predict_expression(face_roi):
    final_image = cv2.resize(face_roi, (224, 224))
    final_image = final_image / 255.0
    final_shape = np.expand_dims(final_image, axis=0)
    
    # Make a prediction
    prediction = new_model.predict(final_shape)
    
    # Get the predicted class index and label
    predicted_class_index = np.argmax(prediction)
    predicted_class_label = class_labels[predicted_class_index]
    
    # Calculate prediction percentage accuracy
    prediction_percentage = prediction[0][predicted_class_index] * 100
    
    return predicted_class_label, prediction_percentage, prediction[0]

def capture_live_image():
    cap = cv2.VideoCapture(0)
    
    while True:
        ret, frame = cap.read()
        if not ret:
            break

        cv2.imshow('Press Space to Capture', frame)
        if cv2.waitKey(1) & 0xFF == ord(' '):
            cap.release()
            cv2.destroyAllWindows()
            return frame

    cap.release()
    cv2.destroyAllWindows()
    return None

def upload_image():
    image_path = input("Enter the path to the image: ")
    image = cv2.imread(image_path)
    if image is None:
        print("Error loading image.")
    return image

def process_and_predict(image):
    if image is not None:
        face_rois, coords = process_faces(image)
        
        for i, face_roi in enumerate(face_rois):
            predicted_class_label, prediction_percentage, prediction = predict_expression(face_roi)
            x, y, w, h = coords[i]
            
            cv2.rectangle(image, (x, y), (x+w, y+h), (255, 0, 0), 2)
            text = f"{predicted_class_label}: {prediction_percentage:.2f}%"
            cv2.putText(image, text, (x, y-10), cv2.FONT_HERSHEY_SIMPLEX, 0.9, (0, 255, 0), 2)
            
            # Print detailed prediction results in the console
            print(f"Prediction: {prediction}")
            print(f"\033[1mPredicted class: {predicted_class_label}\033[0m")
            print(f"Prediction accuracy: {prediction_percentage:.2f}%")
            
            # Print probabilities for all classes
            y_offset = y + h + 20  # Initial y offset for the text
            for idx, prob in enumerate(prediction):
                label = class_labels[idx]
                prob_text = f"{label}: {prob * 100:.2f}%"
                if idx == np.argmax(prediction):
                    # Highlight the predicted class in bold
                    cv2.putText(image, prob_text, (x, y_offset), cv2.FONT_HERSHEY_SIMPLEX, 0.6, (0, 255, 255), 2)
                    print(f"\033[1m{prob_text}\033[0m")
                else:
                    cv2.putText(image, prob_text, (x, y_offset), cv2.FONT_HERSHEY_SIMPLEX, 0.6, (255, 255, 255), 1)
                    print(prob_text)
                y_offset += 20  # Increase y offset for the next class probability

        plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
        plt.title("Image with Prediction")
        plt.show()
    else:
        print("No valid image to process.")

def main():
    choice = input("Choose an option: (1) Capture live image (2) Upload image: ")

    if choice == '1':
        image = capture_live_image()
        process_and_predict(image)
    elif choice == '2':
        image = upload_image()
        process_and_predict(image)
    else:
        print("Invalid choice. Please choose 1 or 2.")

if __name__ == "__main__":
    main()
```
